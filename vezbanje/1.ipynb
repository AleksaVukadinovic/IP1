# Istrazivanje podataka

Istrazivanje podataka je proces automatskog otkrivanja korisnih informacija u velikom skladistu podataka. Kljucna rec VELIKOM.

Da bi mogli da izvucemo neke korisne informacije iz podataka, podataka moramo da imamo mnogo. Tj. da bi primetili sablone (pouzdano) moramo da immao veliki uzorak.

Obrada podataka se sastoji iz 4 glavna koraka:
- priprema podataka
- priprema tehnika istrazivanja podataka
- evaluacija
- primena rezultata

Imamo dve glavne vrste nadgledanog ucenja:
- **Klasifikacija** je odredjivanje klasa tj. kategorija za date podatke. Tj. slikamo podatke u neuredjeni skup nekih kategorija (mogu biti stringovi, intovi itd.)
- **Regresija** podaci se slikaju u neprikadan niz vrednosti tj ide u R.

Kazemo da je ucenje nadgledano jer imamo date ulaze i izlaze i na osnovu njih treniramo modele, tj. kao da imamo ucitelja koji nadgleda i moze da ispravi resenja. Sustinski se razlikuje nenadgledano ucenje, jedan primer je:
**Klasterovanje** - koje podrazumeva da podatke koji su slicni po nekoj metrici grupisemo u klastere

## Kakvi podaci mogu da budu?

- Npr mogu biti tabelarni - objekti su redovi, a kolone su atributi tog objekta
- Mogu da budu slike - matrica hxw, ako je velicina matrice samo hxw onda je greysacle, ako hocemo da bude u boji onda je bi bila hxwx3
- Grafovi - tu mozemo da radimo klasifikaciju cvorova (npr ljudi mogu biti cvorovi, a grane konekcije u slucaju drustvenih mreza, moze se korisitit i za molekule itd.)
- Tekst - Sta je tekst? -> to je string, ali to nam nije udobno za rad, umesto toga koristimo **TF matricu - term frequency** (ove se moze korisiti ako hocemo da klasifikujemo tekstove da li pripadaju politici, ekonmiji, sportu...). Redovi su dokumenti, a kolone su same reci. Mnogo reci su zajednicke i svuda se ponavljaju, ali imamo keywordove koji se javljaju samo u odredjenim tekstovima, na osnovu toga mozemo da odredimo vrstu clanka. Kada treniramo model, moguce je da nam se keywordove izgube cak i ako su nam najbitniji, da bi izbegli to postoje razlicite transformacije koje imaju za cilj da ocuvaju ovo. Nama najbitnija je **IDF** matrica - pomocu nje mozemo da damo vecu vaznost ovim redjim recima tako da ne budu progutane prilikom treninga. **IDF - Inverse Document Frequencey** - svaka frekvencija se mnozi sa $log(N/n_t)$, gde je N broj dokumenata, a $n_t$ broj dokumenata u kojima se term t javlja -> sve reci ce imati tezinu u rasponu $[0, logN]$.
- Vremenski podaci - video, zvuk, ekg ...
- Prostorni podaci - temperatura, nadmorska visina,
- Prostono-vremenski podaci - merenje temperature svaki dan u svakom gradu u Srbiji...

Poenta je da nece svi algoritmi raditi za sve

## Podela atributa prema operacijama

1. **Imenski (nominalni)** – razlika `=`, `≠`  
   - Primer: `Pera = Pera`, `Pera ≠ Jelena`  
   - Uređivanje leksikografski moguće, ali bez stvarnog smisla  

2. **Redni (ordinalni)** – razlika + uređenost `<, >`  
   - Primer: saradnik < asistent < docent < profesor  

3. **Intervalni** – razlika, uređivanje, aditivnost  
   - Primer: `3.2.2022 < 5.3.2022`, može se računati razlika datuma  

4. **Razmerni (ratio)** – sve operacije uključujući i množenje/deljenje  
   - Primer: udaljenost između gradova  

- 1 i 2 = **kvalitativni atributi**  
- 3 i 4 = **kvantitativni atributi**  

---

## Podela atributa prema vrednostima

- **Diskretni** – konačan ili prebrojivo beskonačan skup vrednosti  
  - *Specijalni slučaj*: binarni atributi  
- **Kontinuirani** – vrednosti iz skupa realnih brojeva  

### Asimetrični (retki) podaci
- Prisustvo nenultih vrednosti je značajno  
- Asimetrični binarni atributi – bitne su nenulte vrednosti
