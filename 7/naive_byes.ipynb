# Naive Bayers

```python
import pandas as pd
```

```python
df = pd.read_csv('./balloons.csv')
```

```python
df.describe()
```

```python
X = df[df.columns[:-1]]
X.shape
```

```python
y = df.iloc[:, -1]
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42, stratify=y)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
model = CategoricalNB()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoder.fit(X_train)
```

```python
encoder.categories_
```

```python
X_train = encoder.transform(X_train)
X_test = encoder.transform(X_test)
```

```python
model = CategoricalNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train, y_train_pred)
```

```python
y_test_pred = model.predict(X_test)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
from sklearn.pipeline import Pipeline
# (transofrmation_1, transofrmation_2, ..., transofrmation_n, classifier)
```

```python
pipe = Pipeline([('ordinal encoder', OrdinalEncoder()), ('classifier', CategoricalNB())])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe['ordinal encoder'].categories_
```

```python
pipe['classifier'].class_count_
```

```python
corpus = ['Chinese Beijing chinese',
          'Chinese Chinese Shanghai',
          'Chinese Macao',
          'Tokyo Japan Chinese']
classes = ['yes', 'yes', 'yes', 'no']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(corpus)
```

```python
vectorizer.fit(corpus)
```

```python
X_train = vectorizer.transform(corpus)
```

```python
type(X_train)
```

```python
X_train.toarray()
```

```python
vectorizer.get_feature_names_out()
```

```python
X_train = pd.DataFrame(X_train.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
X_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, classes)
```

```python
model.classes_
```
